```python
import numpy as np
import matplotlib.pyplot as plt
import awkward as ak
import mplhep as hep
hep.style.use(hep.style.CMS)

import hist as hist2
from coffea import processor
from coffea.analysis_tools import Weights

import warnings
warnings.filterwarnings("ignore", message="Found duplicate branch")
```

```python


def invariant_mass(jet1, e):
    """b+e invariant mass"""
    return np.sqrt(2 * jet1.pt * e.pt * (np.cosh(jet1.eta - e.eta) - np.cos(jet1.phi - e.phi)))
```

```python
class BackgroundProcessor(processor.ProcessorABC):
    def __init__(self):
        
        # here we define an output
        self.make_output = lambda: {
            'sumw': 0.,
            "leading_jet_kin": hist2.Hist(
                hist2.axis.Regular(50, 30, 1000, name="jet_pt", label=r"Jet $p_T$ [GeV]"),
                hist2.axis.Regular(20, -5, 5, name="jet_eta", label=r"Jet $\eta$"),
                hist2.axis.Regular(30, 0, 200, name="jet_mass", label=r"Jet $m$ [GeV]"),
                hist2.storage.Weight(),
            ),
            "met_kin": hist2.Hist(
                hist2.axis.Regular(50, 0, 1000, name="met_pt", label=r"$p_T^{miss}$ [GeV]"),
                hist2.storage.Weight(),
            ),
            'minv': hist2.Hist(

                hist2.axis.Regular(30, 0, 5000, name='invariantMass', label=r"$m_{jj}^{max}$ [GeV]"))
        }

    def process(self, events):
        dataset = events.metadata['dataset']
        weights = Weights(len(events), storeIndividual=True)
        weights.add('genweight', events.genWeight)
        
        output = self.make_output()
        output['sumw'] = ak.sum(events.genWeight)
        
        #electron 

        electron=events.Electron

        one_electron=ak.sum(electron.pt>0, axis=1)==1


        # jets
        jets = events.Jet
        
        # select events with at least two jets
        one_jet=ak.sum(jets.pt>0, axis=1)==1

        maks_final=one_jet & one_electron


        electron_jets=events[maks_final]

                
            
        # jets four vectors
        electron_jets_events = ak.zip(
            {
                "pt_eletron": electron_jets.Electron.pt,
                "eta_eletron": electron_jets.Electron.eta,
                "phi_eletron": electron_jets.Electron.phi,
                "mass_eletron": electron_jets.Electron.mass
            }
        )
        # get combinations of jet pairs
        electron_events = ak.zip(
            {
                "pt": electron_jets.Electron.pt,
                "eta": electron_jets.Electron.eta,
                "phi": electron_jets.Electron.phi,
                "mass": electron_jets.Electron.mass
            }
        )

        jet_events = ak.zip(
                    {
                        "pt": electron_jets.Jet.pt,
                        "eta": electron_jets.Jet.eta,
                        "phi": electron_jets.Jet.phi,
                        "mass": electron_jets.Jet.mass
                    }
                )
        
        # compute the invariant mass for every pair of jets
        # and select the maximum value
        max_mjj = ak.max(invariant_mass(jet_events, electron_events), axis=1)
        
        # met
        met = events.MET
        met = met[maks_final]
        

                
        # max delta eta between jets
        max_delta_eta = abs(ak.max(jet_events.eta - electron_events.eta, axis=1))

        def normalize(val, cut=None):
            if cut is None:
                ar = ak.to_numpy(ak.fill_none(val, np.nan))
                return ar
            else:
                ar = ak.to_numpy(ak.fill_none(val[cut], np.nan))
                return ar
    
        # here we fill our histograms

        output["met_kin"].fill(
            met_pt=normalize(met.pt)
        )
        
        output['minv'].fill(
            invariantMass = normalize(max_mjj)
        )

        return {dataset: output}
            
    def postprocess(self, accumulator):
        return accumulator


```
